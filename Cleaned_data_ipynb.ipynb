```python
df.rename(columns={'What is your age? ':'Age','  What is your gender?  ':'Gender','Current Stress Level':'Stress_level',
                   '  Factors Contributing to Stress   [Work-related stress]':'Work_stress'},inplace=True)
```

```python
df.rename(columns={'  Factors Contributing to Stress   [Financial stress]':'Financial_stress',
                  '  Factors Contributing to Stress   [Health issues]':'Health_issues','  Factors Contributing to Stress   [Family and relationship problems]':'Family_problems',
                   '  Factors Contributing to Stress   [Social and environmental factors]':'Social_stress','  Factors Contributing to Stress   [Educational stress]':'Educational_stress',
                   '  Factors Contributing to Stress   [Life changes and uncertainty]':'Life_changes','  Factors Contributing to Stress   [Technology and information overload]':'Technology_overload',
                   '  Factors Contributing to Stress   [Cultural and societal pressures]':'Cultural_stress','  Factors Contributing to Stress   [Personal habits and lifestyles]':'Personal_habits',
                   'What is your country of residence?':'Country','What is the Province of your residence? ':'Province'},inplace=True)
```

```python
df.head()
```

```python
df.drop(columns=['Province'],inplace=True)
```

```python
df.head()
```

```python
df['Country'].str.strip()
```

```python
df.isin(['Nugegoda ']).any()
```

```python
#Replacing Nugegoda with SriLanka hence Nugegoda is a city in SriLanka
df['Country'].replace('Nugegoda ','Sri Lanka',inplace=True)
df.head()
```

```python
#Capitalizing the first letter of each word in the Country column
df['Country']=df['Country'].str.title()
df.tail()
```

```python
def extract_numeric(value):
  if isinstance(value, str):
    return int(value.split()[0])
  else:
    return value


for col in df.columns[4:14]:
  df[col] = df[col].apply(extract_numeric)

df.head()
```
